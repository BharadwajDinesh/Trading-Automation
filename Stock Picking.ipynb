### Get last 31 days data from TradingView for each stock
### Compare 1 month avg Vol. with current days Vol.
### Case 1 if avg_1_month_vol < day_vol the Stock will be displayed

```python
for i in range(len(symbols)):
    
    data = tv.get_hist(symbols[i],'NSE',interval=Interval.in_daily,n_bars=31)
    avg_1_month_vol = data['volume'].iloc[:len(data['volume'])-1].mean() # Calculating 1 month avg Volume
    day_vol = data.iloc[-1]['volume'] # Current day's Volume
    if avg_1_month_vol < day_vol: # Checks if current day's Vol. is greater than 1 month avg Vol.
        print(symbols[i])
    else:
        continue
```

### Case 2 : The stock reaches a volume at least 30% higher
### than it reaches any 9:15-10:30 on an average in the last 30 days
### but its price has not changed more than 2% since it opened

```python
for i in range(len(symbols)):
    
    # Get last 31 days 15min data
    fifteen_min_data = tv.get_hist(symbols[i],'NSE',interval=Interval.in_15_minute,n_bars=775)
    
    # Get today's date
    today_date = str(datetime.today().date())
    
    # Filter rows within the specified time frame (9:15 to 10:30) and filter current day's data
    filtered_df = fifteen_min_data[~(fifteen_min_data.index.date == pd.to_datetime(today_date).date())]
    selected_rows = filtered_df.between_time('09:15:00', '10:30:00')
    
    # 30 days average calculated
    avg_30 = selected_rows['volume'].mean()
    
    # Calculating more than 30% of 30 days avg.
    more_than_30_percent = avg_30 + (0.3 * avg_30)
    
    # Current day's vol.
    today_data = tv.get_hist(symbols[i],'NSE',interval=Interval.in_daily,n_bars=1)
    today_vol = today_data['volume'][0]
    
    # Current day's percentage change.
    today_perc_change = (today_data['close'][0] - today_data['open'][0]) / today_data['open'][0] * 100
    
    if (more_than_30_percent < today_vol) & (today_perc_change < 2): # Checks if current day's Vol. is greater than 30% of month avg Vol.
        print(symbols[i])
    else:
        continue
    
```
